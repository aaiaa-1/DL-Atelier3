## Scraping Booking website : https://www.booking.com/

```python
import urllib.request,urllib.parse,urllib.error
from bs4 import BeautifulSoup
from langdetect import detect
```

```python
def bookingReviews(lien,nbr):
    arabe=[]
    lines=[]
    
    num=(nbr-1)*10
    for i in range(0,num,10):
        # print(i)
        url=lien+str(i)
        print(url)
        req = urllib.request.Request(
        url,
        headers={
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36',
        }
        )
        f = urllib.request.urlopen(req)
        soup = BeautifulSoup(f.read().decode('utf-8'),'html.parser')
        divs = soup.findAll("li", {"class": "review_list_new_item_block"})
        # k=0
        for div in divs:
            commentaire=div.find("h3",{"class":"c-review-block__title--original"}) 
            if commentaire==None :
                continue
            try:
                lang=detect(commentaire.get_text())
            except:
                continue
            if lang!='ar':
                continue
            rating=div.find("div",{"class":"bui-review-score__badge"})
            # print(commentaire.get_text())
            # print(rating.get_text())
            arabe.append((commentaire.get_text(),rating.get_text()))
            
    return arabe
```
